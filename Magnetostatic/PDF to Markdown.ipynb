```python
import fitz
import markdownify

def pdf_to_markdown(pdf_path):
    doc = fitz.open(pdf_path)
    markdown_text = ""
    for page in doc:
        text = page.get_text("text")
        markdown_text += markdownify.markdownify(text) + "\n"
    return markdown_text

if __name__ == "__main__":
    pdf_file = "C:\EMSolution\EMSolpy4\python\Magnetostatic\magnetostatic.pdf"  # PDFファイルのパスを指定
    markdown = pdf_to_markdown(pdf_file)
    print(markdown)
```
